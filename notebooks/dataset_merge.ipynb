```python
import pandas as pd

altor_locations_df = pd.read_excel("Altor Locations.xlsx")

altor_locations_df
```

```python
altor_locations_df = altor_locations_df.rename(columns={
    "Location":"location", "Latitude":"latitude", "Longitude":"longitude"
    })
altor_locations_df
```

```python
altor_locations_df["company"] = "Altor Solutions"
altor_locations_df = altor_locations_df[["company","location" ,"latitude", "longitude"]]

altor_locations_df
```

```python
competitors_location_df = pd.read_csv("Comp Analysis- BI Report.csv")
competitors_location_df = competitors_location_df[["top_competitors", "city", "state", "latitude", "longitude"]]
competitors_location_df = competitors_location_df.rename(columns={"top_competitors":"company"})
```

```python
location = (competitors_location_df.city + ", " + competitors_location_df.state).to_list()
location
```

```python
competitors_location_df["location"] = pd.Series(location)
competitors_location_df["location"]
```

```python
all_companies_df = pd.concat([altor_locations_df, competitors_location_df]).drop(columns=["city", "state"])

all_companies_df.to_csv("companies_locations.csv", index=False)
```
